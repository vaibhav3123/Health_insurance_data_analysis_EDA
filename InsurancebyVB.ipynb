```python
ax = data['age'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black', alpha=0.7, figsize=(8, 5))

for patch in ax.patches:
    height = patch.get_height()
    ax.annotate(f'{int(height)}', 
                (patch.get_x() + patch.get_width() / 2, height), 
                ha='center', va='bottom', fontsize=10)

plt.title('Age Distribution with Counts')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python
data['bmi'].plot(kind = 'hist')
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequncy")
plt.show()
```

```python
x = data['region'].value_counts().plot(kind='bar')
for bars in x.containers:
    x.bar_label(bars)
plt.title("Bar chart that showing region")
plt.xlabel('region')
plt.ylabel('frequency')
plt.show()  
    
```
